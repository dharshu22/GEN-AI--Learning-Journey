```python
import numpy as np
print("Multiplication matrix:")
p1 = np.array([[3,4],[5,6]])
p2 = np .array([[4,5],[8,9]])
print(np.dot(p1,p2))
```

```python
print(p1 @ p2)
```

```python
print("Transpose of matrix:")
print(p1.T)
print(p2.T)
```

```python
print("Random matrix:")
matrix = np.random.randint(1, 100, (3, 3))print("Random Matrix:")
print(matrix)
```

```python
print("Random - Toss a Coin :")
Coin = np.random.choice(["Head","Tail"])
print(Coin)
```

```python
print("Random - Dice Roll:")
dice = np.random.randint(1, 7)
print("Dice Roll:", dice)
```

```python
r1 = np.array([1,2,3])
r2 = np.array([4,5,6])
print("vstack-veritcal stack:")
print(np.vstack((r1,r2)))
```

```python
print("hstack-Horizontal Stack:")
print(np.hstack((r1,r2)))
```

```python
print("Concatenate- Joins two or more arrays:")
print(np.concatenate((r1, r2)))
```

```python
print("split-divides a array into smaller array & arange-evenly spaced array:")
print(np.split(np.arange(12), 4))
```

```python
print("unsorted to sorted array:")
employee = np.array([11,12,13,14,15,16,12,14,15,17])
print(np.sort(employee))
```

```python
print("argsort-index order")
print(np.argsort(employee))
```

```python
print("Employee number greater then 12:")
print(np.where(employee>12))
```
