```python
import nltk
nltk.download('punkt')
nltk.download('punkt_tab')
from nltk.tokenize import sent_tokenize

data = '''Hello everyone! I am priya from Chennai. I love travelling and discovering new destinations whenever I get the chance. 
Exploring different places is one of my favorite hobbies and gives me a fresh perspective.'''
print(sent_tokenize(data))
```

```python
from nltk.tokenize import word_tokenize
data1 = '''Hello everyone! I am priya from Chennai. I love travelling and discovering new destinations whenever I get the chance. 
Exploring different places is one of my favorite hobbies and gives me a fresh perspective.'''
print(word_tokenize(data1))
```

```python
from nltk.tokenize import sent_tokenize
data2 = 'PD Bonne après-midi.'
print(sent_tokenize(data2,'french'))
```
