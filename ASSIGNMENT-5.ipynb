```python
def RATROW():
    name='priya'
    for i in range(len(name)):
        for j in range(0,i+1):
            print(name[i],end=' ')
        print()
RATROW()
```

```python
def RATCOL():
    name='priya'
    for i in range(len(name)):
        for j in range(0,i+1):
            print(name[j],end=' ')
        print()
RATCOL()
```
