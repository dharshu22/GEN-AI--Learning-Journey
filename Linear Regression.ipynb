```python
from sklearn.linear_model import LinearRegression
import numpy as np
house_size = np.array([[800], [1000], [1200], [1500], [1800]])
house_price = np.array([40, 50, 60, 75, 90])  
model = LinearRegression()
model.fit(house_size, house_price)
new_size = np.array([[1300], [1600], [2000]])
predicted_price = model.predict(new_size)
for size, price in zip(new_size, predicted_price):
    print(f"House Size: {size[0]} sq.ft -> Predicted Price: ₹{price:.2f} Lakhs")
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
import numpy as np
house_size = np.array([[800], [1000], [1200], [1500], [1800]])
house_price = np.array([40, 50, 60, 75, 90])
X_train, X_test, y_train, y_test = train_test_split(
    house_size, house_price, test_size=0.2, random_state=42)

model = LinearRegression()
model.fit(X_train, y_train)
test_predictions = model.predict(X_test)
print("Test Data Predictions")
print("-" * 35)
for size, actual, predicted in zip(X_test, y_test, test_predictions):
    print(f"House Size: {size[0]} sq.ft")
    print(f"Actual Price: ₹{actual} Lakhs")
    print(f"Predicted Price: ₹{predicted:.2f} Lakhs\n")
new_size = np.array([[1300], [1600], [2000]])
predicted_price = model.predict(new_size)
print("New House Price Predictions")
print("-" * 35)
for size, price in zip(new_size, predicted_price):
    print(f"House Size: {size[0]} sq.ft -> Predicted Price: ₹{price:.2f} Lakhs")
```
