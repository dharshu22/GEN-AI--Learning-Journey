```python
import numpy as np
print("Storing Employee's Salary in 1D and 2D Array:")
p = np.array([3000,5000,4000,2000])
r = np.array([[1000,2000,3000],[4000,5000,6000]])
print(p)
print(r)
```

```python
print("Creats a empty array filled with 0:")
p = np.zeros((3,3))
print(p)
```

```python
print("An array Filled with 1:")
ones = np.ones((3,3))
print(ones)
```

```python
print("Creating an identity matrix:")
identity = np.eye(3)
print(identity)
```

```python
print("Even space between the values:")
ar = np.arange(0,10,2)
print(ar)
```

```python
print("Evenly spaces 5 values:")
lin = np.linspace(0,1,5)
print(lin)
```

```python
print("Attributes that describe its shape,no of dimensions,total number of elements and its data type:")
print(r.shape)
print(r.ndim)
print(r.size)
print(r.dtype)
```

```python
print("Indexing and slicing in array:")
print(r[0, 1])
```

```python
print(" It returns entire column 2 values:")
print(r[:, 2])
```

```python
print("It returns entire  row 1 values:")
print(r[1, :])
```

```python
print("It returns row 0,1 and column 0,1 values:")
print(r[0:2, 0:2])
```
