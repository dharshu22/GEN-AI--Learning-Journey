```python
doc=nlp("A TL is responsible for leading and guiding a team to achieve project goals efficiently. They assign tasks, monitor progress, and ensure  work is completed on time and meets quality standards.")  
len(doc)
```

```python
for i in doc.sents:
    print(i.text)
```

```python
for i in doc:
    print(i.text,'--',i.lemma_)
```

```python
for i in doc:
    print(i.text,'--',i.pos_)
```

```python
for i in doc:
    print(i.text,'--',i.tag_)
```

```python
nouns=[]
adjectives=[]
for i in doc:
    if i.pos_=='NOUN':
        nouns.append(i.text)
print(nouns)
```

```python
for i in doc:
    if i.pos_=='ADJ':
        adjectives.append(i.text)
print(adjectives)
```
