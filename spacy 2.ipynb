```python
doc=nlp("A TL is responsible for leading and guiding a team to achieve project goals efficiently. They assign tasks, monitor progress, and ensure  work is completed on time and meets quality standards.")  
len(doc)
```

```python
def flat_subtree(stmt):
    return''.join([token.text_with_ws for token in list(stmt)]).strip()
print('Sentence:')
print(list(doc.sents)[0])
print(list(doc.sents)[1])
```

```python
for word in list(doc.sents)[1]:
    print('word:',word.text)
    print('POS:',word.pos_)
    print('Dependency:',word.dep_)
    print('flatten-subtree:',flat_subtree(word.subtree))
    print('*'*30)
```
