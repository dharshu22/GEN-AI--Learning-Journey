```python
print('grade of languages using elif stmt')
print('------------------------------------')
course=['tamil','english','maths']
tamil = int(input('enter tamil mark:'))
english = int(input('enter english mark:'))
maths = int(input('enter maths mark:'))

marks=[tamil,english,maths]
for i in range(3):
    mark = marks[i]
    if mark>=91 and mark<=100:
        print(course[i],'s grade',mark)
    elif mark>=81 and mark<=90:
        print(course[i],'A grade',mark)
    elif mark>=71 and mark<=80:
        print(course[i],'B grade',mark)
    elif mark>=61 and mark<=70:
        print(course[i],'C grade',mark)
    elif mark>=51 and mark<=60:
        print(course[i],'D grade',mark)
    elif mark==50:
        print(course[i],'E grade',mark)
    else:
        print(course[i],'Fail',mark)
```

```python
###### print('BMI Calculator')
print('--------------')

weight = float(input('Enter weight (kg): '))
height = float(input('Enter height (m): '))

bmi = weight / (height * height)

print('BMI =', bmi)

if bmi <= 20.4:
    print('Underweight')
elif bmi <= 30:
    print('Normal Weight')
elif bmi <= 40:
    print('Overweight')
else:
    print('Obese')
```
