```python
print('IRAT-NUM-ROW')
for i in range(6,0,-1):
    for j in range(0,i):
        print(i, end=' ')
    print()
    
```

```python
print('IRAT-NUM-COL')
for i in range(6,0,-1):
    for j in range(0,i):
        print(j, end=' ')
    print()
```

```python
print('IRAT-UPPER-ROW')
for i in range(6,0,-1):
    for j in range(0,i):
        print(chr(i+64), end=' ')
    print()
```

```python
print('IRAT-UPPER-COL')
for i in range(6,0,-1):
    for j in range(1,i):
        print(chr(j+64), end=' ')
    print()
```

```python
print('IRAT-LOWER-ROW')
for i in range(6,0,-1):
    for j in range(0,i):
        print(chr(i+64+32), end=' ')
    print()
```

```python
print('IRAT-LOWER-COL')
for i in range(6,0,-1):
    for j in range(1,i):
        print(chr(j+64+32), end=' ')
        print()
```
