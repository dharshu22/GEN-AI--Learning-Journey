```python
from bs4 import BeautifulSoup
import urllib.request
import nltk
from nltk.corpus import stopwords
data = urllib.request.urlopen('https://www.github.com')
content = data.read()
soup = BeautifulSoup(content)
text_data =  soup.get_text(strip = True)
tkn=[i for i in text_data.split()]

clear_tkn = tkn[:]
word = stopwords.words('english')
for t in tkn:
    if t in stopwords.words('english'):
        clear_tkn.remove(t)
        
fd = nltk.FreqDist(clear_tkn)
for i,j in fd.items():
    print(str(i),'=',str(j))
```

```python
fd.plot(25)
```
