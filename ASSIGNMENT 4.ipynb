```python
#LAT-NUM-ROW
for i in range(1,6):
    for j in range(6,i,-1):
        print(' ',end=' ')
    for k in range(0,i):
         print(i,end=' ')
    print()
```

```python
#LAT-NUM-COL
for i in range(1,6):
    for j in range(6,i,-1):
        print(' ',end=' ')
    for k in range(0,i):
         print(k,end=' ')
    print()
```

```python
#LAT-STAR
for i in range(1,6):
    for j in range(6,i,-1):
        print(' ',end=' ')
    for k in range(0,i):
        print('*',end=' ')
    print()
```

```python
#lat-upper-row
for i in range(1,6):
    for j in range(6,i,-1):
        print(' ',end=' ')
    for k in range(0,i):
         print(chr(i+64),end=' ')
    print()
```

```python
#lat-upper-col
for i in range(1,7):
    for j in range(6,i,-1):
        print(' ',end=' ')
    for k in range(1,i):
         print((chr(k+64)),end=' ')
    print()
```
