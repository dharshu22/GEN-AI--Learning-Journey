```python
from nltk.corpus import stopwords
stopwords_list = set(stopwords.words('english'))
for i in stopwords_list:
    print(i)
```

```python
word = "honor"
stop_words = set(stopwords.words("english"))
print("word:",word)
```

```python
from nltk.corpus import wordnet
synsets = wordnet.synsets(word)
synonym_list = []
for synm in synsets:
    for lemma in synm.lemmas():
        if lemma.name() not in synonym_list:
            synonym_list.append(lemma.name())
print(synonym_list)
```

```python
antonym_list = []
for i in wordnet.synsets('honor'):
    for j in i.lemmas():
        if j.antonyms():
            antonym_list.append(j.antonyms()[0].name())    
print(antonym_list)
```

```python
from nltk.stem import PorterStemmer
word = PorterStemmer()
print(word.stem('visualizing'))
```

```python
from nltk.stem import LancasterStemmer
word = LancasterStemmer()
print(word.stem('visualizing'))
```

```python
from nltk.stem import SnowballStemmer
word = SnowballStemmer('english')
print(word.stem('liking'))
```
