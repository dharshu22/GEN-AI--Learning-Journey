```python
import numpy as np
print("Storing Employee's Salary in 1D and 2D Array:")
p = np.array([3000,5000,4000,2000])
r = np.array([[1000,2000,3000],[4000,5000,6000]])
print(p)
print(r)
```

```python
print("Reshape - Change the Shape(Dimensions) without changing its data:")
employee = np.array([100,101,102,103,104,105])
reshaped = employee.reshape(6,1)
print(reshaped)
```

```python
print("converts the original array into row by row:")
flat = r.flatten()
print(flat)
```

```python
raveled = r.ravel()
print(raveled)
```

```python
print("Arithmetic operations:")
x = np.array([6,7,8,9])
y = np.array([1,2,3,5])
print(x + y)
print(x - y)
print(x * y)
print(x / y)
print(x**2)
```

```python
print("A array with different shapes are automatical expands a smaller array:")
matrix = np.array([[11,12,13,14], [15,16,17,18]])
vector = np.array([40,50,60,70])

print(matrix + vector)
```

```python
print("Aggregation Function:")
print(matrix.sum())
print(matrix.mean())
print(np.median(matrix))
print(matrix.std())
print(matrix.var())
print(matrix.min())
print(matrix.max())
print(matrix.sum(axis=0))
print(matrix.sum(axis=1))
```
