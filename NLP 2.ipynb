```python
from bs4 import BeautifulSoup
import urllib.request 
import nltk
data = urllib.request.urlopen('https://www.github.com')
content = data.read()
soup = BeautifulSoup(content)
content_data = soup.get_text(strip=True)
s_data = [i for i in content_data.split()]
fq = nltk.FreqDist(s_data) 
for k,v in fq.items():
    print(str(k),':',str(v))
    
```

```python
fq.plot(25)
```
