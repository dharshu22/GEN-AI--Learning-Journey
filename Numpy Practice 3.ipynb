```python
#Shopping Discount:
import numpy as np
prices = np.array([800, 1200, 950, 1500, 2000])
discount = prices[prices > 1000]
print("Products Eligible for Discount:")
print(discount)
```

```python
#Random data generation - weather forecast:
temperature = np.random.randint(25, 41, 7)
print("Weekly Temperature:")
print(temperature)
```

```python
# sorting- employee salary ranking:
salary = np.array([45000, 30000, 55000, 40000])
sorted_salary = np.sort(salary)
print(sorted_salary)
```

```python
sales = np.array([8500, 6200, 9100, 7400, 5600])
print("Sorted Sales:")
print(np.sort(sales))
```

```python
#Reshape - Class room Arrangement:
students = np.arange(1, 21)
seating = students.reshape(4, 5)
print(seating)
```

```python
passengers = np.arange(1, 31)
seats = passengers.reshape(6, 5)
print(seats)
```

```python
# Matrix Multiplication:
units = np.array([[120],
                  [180],
                  [150]])
cost = 8
bill = units * cost
print("Electricity Bill:")
print(bill)
```
